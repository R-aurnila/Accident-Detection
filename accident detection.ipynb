```python
import os
import numpy as np
import shutil
import random
root_dir = r'C:\Users\Administrator\Desktop\my research(Shakib)\accident detection\dest\\'
classes_dir = ['accident','noaccident']

test_ratio = 0.20
#create folder train and test , every folder have two subfolder -->(accident and noaccident)
for cls in classes_dir:
    os.makedirs(root_dir +'train/' + cls)
    os.makedirs(root_dir +'test/' + cls)
```

```python
import tqdm
def call2(filen):#accident
    #define the src folder
    #sr=r"C:\Users\Administrator\Desktop\my research(Shakib)\accident detection\accident" +accident
    src = os.path.join(sr,filen)
    #read the src folder
    allFileNames = os.listdir(src)
    #randomize the files names
    np.random.shuffle(allFileNames)
    #split the files in train and test
    train_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                              [int(len(allFileNames)* (1 - test_ratio))])
    
    #add the full url
    train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
    test_FileNames = [src+'/' + name for name in test_FileNames.tolist()]

    print("*****************************")
    print('Total images: ', len(allFileNames))
    print('Training: ', len(train_FileNames))
    print('Testing: ', len(test_FileNames))
    print("*****************************")

    #copy the train and test files in corresponding folder
    for name in tqdm.tqdm(train_FileNames):
        try:
            shutil.copy(name, root_dir +'train/'+filen+'/' )
        except:
            pass

    for name in tqdm.tqdm(test_FileNames):
        #if any error it ignores this file.
        try:
            shutil.copy(name, root_dir +'test/'+filen+'/')
        except:
            pass
    print("Copying Done!")
```

```python
for x in classes_dir:
  # we called every class folder from here. --> accident ,noaccident
  call2(x)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
classifier = Sequential()


classifier.add(Conv2D(32, (3, 3), input_shape=(64, 64, 3), activation='relu'))


classifier.add(MaxPooling2D(pool_size=(2, 2)))


classifier.add(Conv2D(32, (3, 3), activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2, 2)))


classifier.add(Flatten())


classifier.add(Dense(units=128, activation='relu'))
classifier.add(Dense(units=1, activation='sigmoid'))


classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1. / 255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1. / 255)

training_set = train_datagen.flow_from_directory(r"C:\Users\Administrator\Desktop\my research(Shakib)\accident detection\dest\train",
                                                 target_size=(64, 64),
                                                 batch_size=64,
                                                 class_mode='binary')

test_set = test_datagen.flow_from_directory(r"C:\Users\Administrator\Desktop\my research(Shakib)\accident detection\dest\test",
                                            target_size=(64, 64),
                                            batch_size=32,
                                            class_mode='binary')

classifier.fit_generator(training_set,
                         steps_per_epoch=96,
                         epochs=100,
                         validation_data=test_set,
                         validation_steps= 13 )
```

```python
from keras_preprocessing import image
import tensorflow as tf
```

```python
def predict(image_path):
    test_image = tf.keras.preprocessing.image.load_img(image_path, target_size=(64, 64))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis=0)
    result = model.predict(test_image)
#     print(training_set.class_indices)
#     print(result)
    if int(result[0][0]) == 1:
        print("no accident")
    else:
        print("accident")
    return int(result[0][0])
```
